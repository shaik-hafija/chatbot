```python
from ktrain.text.summarization import TransformerSummarizer
ts = TransformerSummarizer()
```

```python
val=df['captions'].iloc[:10]
val
for i in val:
  print(i)
```

```python
summarized=[]
for i in val:
  resu=ts.summarize(i)
  print(resu)
  summarized.append(resu)
```

```python
dfnew = pd.DataFrame(summarized)
dfnew
```

```python
!pip install rouge
```

```python

from rouge import Rouge
rouge = Rouge()

for i in range(len(summarized)):
  scores = rouge.get_scores(val[i], summarized[i])
  print(scores)
```
